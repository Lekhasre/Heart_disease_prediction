```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"D:\Ml projects\Heart Disease\dataset.csv")
```

```python
x = data.drop('target', axis = 1)
y = data['target']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = GradientBoostingClassifier(n_estimators = 300, learning_rate = 0.05, max_depth = 3, random_state = 42)
```

```python
model.fit(x_train.values, y_train.values)
```

```python
x_train_prediction = model.predict(x_train.values)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Training data Accuracy:", training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test.values)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print("Testing data Accuracy:", test_data_accuracy)
```

```python
input = (44,1,2,150,288,0,0,150,1,3.0,2)
input_as_numpy_arr = np.asarray(input)
input_reshaped = input_as_numpy_arr.reshape(1,-1)
prediction = model.predict(input_reshaped)
print(prediction)

if(prediction[0]==0):
    print("The person doesn't have Heart Disease!")
else:
    print("The person has Heart Disease!")
```
